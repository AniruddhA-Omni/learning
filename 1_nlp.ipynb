```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from rich import print
```

```python
stemming = PorterStemmer()

words = ["running", "ran", "runner", "easily", "fairly", "eat", "eating", "ate", "eaten", "better", "best", "worse", "worst"]

for word in words:
    stemmed_word = stemming.stem(word)
    print(f"Original: {word} -> Stemmed: {stemmed_word}")
```

```python
from nltk.stem import RegexpStemmer

reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)

words = ["running", "ran", "runner", "easily", "fairly", "eat", "eating", "ate", "eaten", "better", "best", "worse", "worst"]

for word in words:
    stemmed_word = reg_stemmer.stem(word)
    print(f"Original: {word} -> Stemmed: {stemmed_word}")
```

```python
from nltk.stem import SnowballStemmer

snowball_stemmer = SnowballStemmer("english")

words = ["running", "ran", "runner", "easily", "fairly", "eat", "eating", "ate", "eaten", "better", "best", "worse", "worst"]

for word in words:
    stemmed_word = snowball_stemmer.stem(word)
    print(f"Original: {word} -> Stemmed: {stemmed_word}")
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

words = ["running", "ran", "runner", "easily", "fairly", "eat", "eating", "ate", "eaten", "better", "best", "worse", "worst"]

for word in words:
    lemmatized_word = lemmatizer.lemmatize(word, pos='v')  # Specify the part of speech as verb
    print(f"Original: {word} -> Lemmatized: {lemmatized_word}")
```

```python
# POS tagging example
text = ["The quick brown fox jumps over the lazy dog", "I am learning natural language processing", "Python is a great programming language"]

for sentence in text:
    tokens = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens)
    print(f"Sentence: {sentence}")
    print(f"POS Tags: {pos_tags}")
    print()
```

```python
# Named Entity Recognition (NER) example
from nltk import ne_chunk
words = ["Barack Obama was the 44th President of the United States.", "Apple Inc. is a technology company headquartered in Cupertino, California.", "The Eiffel Tower is located in Paris, France."]

for sentence in words:
    tokens = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens)
    named_entities = ne_chunk(pos_tags)
    print(f"Sentence: {sentence}")
    print(f"Named Entities: {named_entities}")
    print()
```
